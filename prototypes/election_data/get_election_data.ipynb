```python
#  get ballots

# read the uuids we already downloaded
with open(f'data/done_uuids.json') as f:
    json_data = f.read()
done_uuids = json.loads(json_data)
i = len(done_uuids)
print(f'already downloaded {i} ballots')

for uuid in uuids:
    if not(uuid in done_uuids):
        url = ELECTION_URL + '/ballots/' + uuid + '/last'
        response = requests.get(url)

        if response.status_code == 200:

            # Extract the response data as a JSON object
            ballot_data = response.json()

            # Open a new file in write mode
            with open(f'data/ballots/{uuid}.json', 'w') as f:

                # Write the data to the file in human-readable JSON format
                json.dump(ballot_data, f, indent=4)
            print(f'ballot {i} saved, uuid: {uuid}')
            done_uuids.append(uuid)

            # save to a file downloaded uuids
            with open(f'data/done_uuids.json', 'w') as f:
                json.dump(done_uuids, f, indent=4)

        else:
            print('Error:', response.status_code)
        i = i+1

print(f'already downloaded {len(done_uuids)} ballots')
```

```python
# get trustees
url = ELECTION_URL + 'trustees'
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the response data as a JSON object
    trustees_data = response.json()

    with open('data/trustees.json', 'w') as f:
        # Write the data to the file in human-readable JSON format
        json.dump(trustees_data, f, indent=4)
    print('trustees saved to trustees.json')
else:
    # Print an error message if the request was not successful
    print('Error:', response.status_code)
```

```python
# get results
url = ELECTION_URL + 'result'
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the response data as a JSON object
    result_data = response.json()

    with open('data/result.json', 'w') as f:
        # Write the data to the file in human-readable JSON format
        json.dump(result_data, f, indent=4)
    print('result saved to result.json')
else:
    # Print an error message if the request was not successful
    print('Error:', response.status_code)
```

```python
# analyse ballot structure
uuid = "0a3ab828-040e-490c-b12b-01575ccfbf71"
with open(f'data/ballots/{uuid}.json') as f:
    json_data = f.read()

# Convert the JSON data into a Python object
ballot = json.loads(json_data)

print(f'ballot keys = {list(ballot.keys())}')
print(f'vote keys = {list(ballot["vote"].keys())}')
print(f'answers length = {len(list(ballot["vote"]["answers"]))}')
answers = list(ballot["vote"]["answers"])
for i in range(len(answers)):
    a = answers[i]
    print(f'answer {i+1} has:')
    print(f'    {len(a["choices"])} choises')
    print(f'    {len(a["individual_proofs"])} individual_proofs')
    if a["overall_proof"]:
        print(f'    {len(a["overall_proof"])} overall_proof')
```
